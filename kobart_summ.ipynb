```python
import torch
from transformers import PreTrainedTokenizerFast
from transformers import BartForConditionalGeneration

tokenizer = PreTrainedTokenizerFast.from_pretrained('gogamza/kobart-summarization')
model = BartForConditionalGeneration.from_pretrained('gogamza/kobart-summarization')
```

```python
import pandas as pd
df_submit = pd.read_csv("/content/drive/MyDrive/GPT/sample_submission.csv")
df_txt = pd.read_csv("/content/drive/MyDrive/GPT/test.csv")
```

```python
# txt 만 뽑아오기기
txt = df_txt.text
ex_text = txt[0]
# tokenizer
input_ids = tokenizer.encode(ex_text)
```

```python
# 문장의 시작과 끝 토큰 추가하기
input_ids = [tokenizer.bos_token_id] + input_ids + [tokenizer.eos_token_id]
input_ids = torch.tensor([input_ids])
```

```python
input_ids
```

```python
summary_txt_ids = model.generate(input_ids = input_ids,
                                 bos_token_id = model.config.bos_token_id,
                                 eos_token_id = model.config.eos_token_id,
                                 length_penalty=2.0,
                                 max_length = 128,
                                 min_length = 32,
                                 num_beams = 4)
print(summary_txt_ids)
```

```python
print(tokenizer.decode(summary_txt_ids[0], skip_special_tokens = True ))
```

```python
summary_txt_ids2 = model.generate(input_ids = input_ids,
                                 bos_token_id = model.config.bos_token_id,
                                 eos_token_id = model.config.eos_token_id,
                                 length_penalty=1.0,
                                 max_length = 128,
                                 min_length = 32,
                                 num_beams = 4)
print(tokenizer.decode(summary_txt_ids2[0], skip_special_tokens = True ))
```

```python
text_list = []
for i, text in enumerate(txt):
  input_ids = tokenizer.encode(text)
  input_ids = [tokenizer.bos_token_id] + input_ids + [tokenizer.eos_token_id]
  input_ids = torch.tensor([input_ids])

  summary_txt_ids = model.generate(input_ids = input_ids,
                                  bos_token_id = model.config.bos_token_id,
                                  eos_token_id = model.config.eos_token_id,
                                  length_penalty=1.0,
                                  max_length = 128,
                                  min_length = 32,
                                  num_beams = 4)

  text_list.append(tokenizer.decode(summary_txt_ids[0], skip_special_tokens = True ))
  print(i)
```

```python
# 결과확인 
df_submit.summary = text_list
df_submit.head()

df_txt.text = text_list
df_txt.head()
```
