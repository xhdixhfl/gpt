```python
from transformers import PreTrainedTokenizerFast, BartForConditionalGeneration
#  Load Model and Tokenize
tokenizer = PreTrainedTokenizerFast.from_pretrained("ainize/kobart-news")
model = BartForConditionalGeneration.from_pretrained("ainize/kobart-news")
```

```python
import pandas as pd
df =pd.read_csv('/content/drive/MyDrive/GPT/test.csv')
df.head()
```

```python
# Encode Input Text
input_text = df.text[0]
input_ids = tokenizer.encode(input_text, return_tensors="pt")
# Generate Summary Text Ids
summary_text_ids = model.generate(
    input_ids=input_ids,
    bos_token_id=model.config.bos_token_id,
    eos_token_id=model.config.eos_token_id,
    length_penalty=1.0,
    max_length=142,
    min_length=56,
    num_beams=4,
)
# Decoding Text
print(tokenizer.decode(summary_text_ids[0], skip_special_tokens=True))
```

```python
text = df.text
txt_list = []
```

```python
for i, t in enumerate(text):
  input_ids = tokenizer.encode(t, return_tensors="pt")
# Generate Summary Text Ids
  summary_text_ids = model.generate(
    input_ids=input_ids,
    bos_token_id=model.config.bos_token_id,
    eos_token_id=model.config.eos_token_id,
    length_penalty=1.0,
    max_length=142,
    min_length=56,
    num_beams=4,
)
# Decoding Text
  txt_list.append(tokenizer.decode(summary_text_ids[0], skip_special_tokens=True))
  print(i)
```

```python
# 결과 확인
txt_list[:3]
```

```python
# 데이터 저장할 파일 불러오기
dx = pd.read_csv('/content/drive/MyDrive/GPT/sample_submission.csv')

# 결과값 채우기
dx.summary = txt_list
dx.head() # 확인
```
