- CONCAT 할 때, 인덱스가 다르거나 개수가 맞지 않으면, 진행 불가. 저장할때 인덱스저장을 안하여서 인덱스 순서가 달랐음. 그를 해결하고자 여러개를 손보다가 오히려 1번행을 삭제 시키는 바람에 1행을 다시 만들고자 하였고, 번호가 잘못된 인덱스를 고치고자 RangeIndex를 사용하여 올바른 인덱스로 변형함. 그후 concat을 진행하여 저장 형식을 맞춰줌

```python
result.index = pd.RangeIndex(start=0, stop=500, step=1)
result
```
